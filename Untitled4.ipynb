### 2.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
x,y=load_iris(return_X_y=True)
from sklearn.svm import SVC
svc=SVC(kernel='poly')
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
svc.fit(x_train,y_train)
```

```python
svc.score(x_test,y_test)
```

### 3.

In Support Vector Regression (SVR), increasing the value of epsilon (epsilon parameter) in the epsilon-SVR model allows for a wider margin of tolerance around the predicted values. This means that the SVR model accepts a certain amount of deviation from the actual target values. As epsilon increases, more data points may fall within the acceptable margin, potentially leading to a decrease in the number of support vectors.

### 4.

The choice of kernel function, C parameter, epsilon parameter, and gamma parameter in Support Vector Regression (SVR) can significantly affect the performance of the model:

Kernel Function:

Different kernel functions (linear, polynomial, radial basis function, etc.) capture different types of relationships in the data. The choice depends on the characteristics of your data.
C Parameter:

The C parameter controls the trade-off between smoothness of the solution and fitting the training data. Smaller values of C result in a smoother decision boundary, while larger values allow the model to fit the training data more closely.
Epsilon Parameter:

The epsilon parameter determines the width of the epsilon-tube around the predicted values. It is crucial for controlling the tolerance of the model to errors.
Gamma Parameter:

The gamma parameter defines the influence of a single training example, with low values meaning far influence and high values meaning close influence. It plays a role in determining the flexibility of the decision boundary.
Here are examples of when you might adjust these parameters:

Increase C when you want a more complex model that fits the training data closely.
Adjust epsilon based on how much deviation from the actual values is acceptable in your problem.
Experiment with different kernel functions based on the underlying patterns in your data.
Adjust gamma to control the influence of individual training examples.

### 5.

```python
y_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,r2_score
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

#defining parameter range.

paramater={'C':[0.1,1,10,100,1000],
          'gamma':[1,0.1,0.01,0.001,0.0001],
           'kernel':['linear','rbf','poly','sigmoid']
          }
grid=GridSearchCV(SVC(),param_grid=paramater,refit=True,cv=5,verbose=3)
grid.fit(x_train,y_train)
```

```python
y_pred2=grid.predict(x_test)
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
print(classification_report(y_test,y_pred2))
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```
